```python
import ee
# Trigger the authentication flow.
ee.Authenticate()
# Initialize the library.
ee.Initialize()
```

```python
col = ee.ImageCollection('LANDSAT/LC08/C02/T1_L2')
```

```python
u_lon = 4.8148
u_lat = 45.7758
u_poi = ee.Geometry.Point(u_lon, u_lat)
```

```python
i_date = '2017-01-01'
f_date = '2020-01-01'
```

```python
scale=20
```

```python
LS = col.filterDate(i_date, f_date)
```

```python
LS=LS.getRegion(u_poi,scale ).getInfo()
```
