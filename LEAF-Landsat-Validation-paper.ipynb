```python
def agregate_estimates(indata,operator='median'):
    if operator=='median':
        outdata=indata.median(axis=0, skipna=True, numeric_only=True)
    elif operator=='mean':
        outdata=indata.mean(axis=0, skipna=True, numeric_only=True)
    else:
        ValueError
    return outdata      
```

```python
def features2DF(Dictionary):
    TAB=None
    for ftr in list(Dictionary.keys()):
        for idx in range(len(Dictionary[ftr])):
            cols=list(Dictionary[ftr][idx].keys()) 
            row=Dictionary[ftr][idx]['feature']
            est=Dictionary[ftr][idx][cols[1]]
            if not est.empty:
                est=agregate_estimates(Dictionary[ftr][idx][cols[1]]).to_dict()
                row.update(est)
                row_df=pd.DataFrame.from_dict([row], orient='columns')
                if TAB is not None:
                    TAB=pd.concat([TAB,row_df], ignore_index=True)
                else: 
                    TAB=row_df             
    return TAB         
```

```python
def format_feature(feat,variablename):
    D=dict()
    for key in feat.keys():
        cols=list(feat[key][0].keys())
        A=[feat[key][ii][cols[0]] for ii in range(len(feat[key]))]
        B=[feat[key][ii][cols[1]] for ii in range(len(feat[key]))]
        for ii in range(len(feat[key])):
            A[ii]['sample%s'%(variablename)]=B[ii]
        D.update({key:A}) 
    return D
```
